```python
# Python built-in library for RegEx
import re
```

```python
sentence = "C:\desktop\notes"
print(sentence)
```

```python
# r meaning threating the string as raw string
# special charectors such as `\n` will be consider as the string `\n` instead of new row.
sentence = r"C:\desktop\notes"
print(sentence)
```

```python
result_search = re.search("C:", sentence)
print(result_search)
```
